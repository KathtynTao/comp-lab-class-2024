```python
# Import necessary libraries
import os
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.analysis import rdf

# Define paths
equil_path = "/home/st5430/comp-lab-class/comp-lab-class-2024/Week9-LAMMPS-SupercooledLiquids/Data/equil"
prod_path = "/home/st5430/comp-lab-class/comp-lab-class-2024/Week9-LAMMPS-SupercooledLiquids/Data/prod"
figure_path = "/home/st5430/comp-lab-class/comp-lab-class-2024/Week9-LAMMPS-SupercooledLiquids/Figures"
os.makedirs(figure_path, exist_ok=True)

# Dictionary of topology and trajectory files, keyed by temperature
files = {
    0.475: ("kalj_T0.475_n360_v300_30000000_1.lammpstrj", "kalj_T0.475_n360_v300_prod_30000000_1.unwrap.dcd"),
    0.5: ("kalj_T0.5_n360_v300_10000000_1.lammpstrj", "kalj_T0.5_n360_v300_prod_10000000_1.unwrap.dcd"),
    0.55: ("kalj_T0.55_n360_v300_3000000_1.lammpstrj", "kalj_T0.55_n360_v300_prod_3000000_1.unwrap.dcd"),
    0.6: ("kalj_T0.6_n360_v300_1000000_1.lammpstrj", "kalj_T0.6_n360_v300_prod_1000000_1.unwrap.dcd"),
    0.65: ("kalj_T0.65_n360_v300_350000_1.lammpstrj", "kalj_T0.65_n360_v300_prod_350000_1.unwrap.dcd"),
    0.7: ("kalj_T0.7_n360_v300_180000_1.lammpstrj", "kalj_T0.7_n360_v300_prod_180000_1.unwrap.dcd"),
    0.8: ("kalj_T0.8_n360_v300_150000_1.lammpstrj", "kalj_T0.8_n360_v300_prod_150000_1.unwrap.dcd"),
    0.9: ("kalj_T0.9_n360_v300_150000_1.lammpstrj", "kalj_T0.9_n360_v300_prod_150000_1.unwrap.dcd"),
    1.0: ("kalj_T1_n360_v300_50000_1.lammpstrj", "kalj_T1_n360_v300_prod_50000_1.unwrap.dcd"),
    1.5: ("kalj_T1.5_n360_v300_10000_1.lammpstrj", "kalj_T1.5_n360_v300_prod_10000_1.unwrap.dcd")
}

msd_results = {}
rdf_results = {(1, 1): [], (2, 2): [], (1, 2): []}  # Store RDF results

# Function to manually calculate MSD
def calculate_msd(universe, selection):
    n_frames = len(universe.trajectory)
    selected_atoms = universe.select_atoms(selection)
    init_positions = selected_atoms.positions.copy()
    msd_values = []

    for ts in universe.trajectory:
        displacements = selected_atoms.positions - init_positions
        squared_displacements = np.sum(displacements ** 2, axis=1)
        msd_values.append(np.mean(squared_displacements))

    return np.arange(n_frames) * universe.trajectory.dt, msd_values

# Compute MSD and RDF for each temperature
for temp, (top_file, traj_file) in files.items():
    top_path = os.path.join(equil_path, top_file)
    traj_path = os.path.join(prod_path, traj_file)

    if not os.path.exists(top_path) or not os.path.exists(traj_path):
        print(f"File not found: {top_path} or {traj_path}")
        continue

    u = mda.Universe(top_path, traj_path, topology_format="LAMMPSDUMP")

    # Calculate MSD
    times, msd_data = calculate_msd(u, 'type 1')
    msd_results[temp] = (times, msd_data)

    # Calculate RDF
    for pair, label in zip([(1, 1), (2, 2), (1, 2)], ["g_11", "g_22", "g_12"]):
        rdf_analysis = rdf.InterRDF(
            u.select_atoms(f"type {pair[0]}"),
            u.select_atoms(f"type {pair[1]}"),
            nbins=200,
            range=(0.1, 10.0)
        )
        rdf_analysis.run()
        rdf_results[pair].append((temp, rdf_analysis.results.bins, rdf_analysis.results.rdf))

# Plot and save MSD for all temperatures
plt.figure()
for temp, (times, msd_data) in msd_results.items():
    plt.loglog(times, msd_data, label=f"T = {temp}")
plt.xlabel("Time (ps)")
plt.ylabel("MSD (nm²)")
plt.title("Mean Squared Displacement at Different Temperatures")
plt.legend()
plt.savefig(f"{figure_path}/MSD_all_temperatures.png")
plt.show()

# Plot and save RDF for all temperatures
for pair, label in zip([(1, 1), (2, 2), (1, 2)], ["g_11", "g_22", "g_12"]):
    plt.figure()
    for temp, bins, rdf_data in rdf_results[pair]:
        plt.plot(bins, rdf_data, label=f"T = {temp}")
    plt.xlabel("Distance (nm)")
    plt.ylabel("g(r)")
    plt.title(f"Radial Distribution Function {label}(r) at Different Temperatures")
    plt.legend()
    plt.savefig(f"{figure_path}/RDF_{label}_all_temperatures.png")
    plt.show()

print("Analysis complete. MSD and RDF plots for all temperatures have been saved to the Figures folder.")
```
