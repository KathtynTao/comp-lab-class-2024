```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Function to load energy data from .ener files
def load_energy_data(file_path):
    return np.loadtxt(file_path, comments='@', skiprows=24)

# Load potential energy data for the METADYN (QM/MM) and subsequent simulations
metadyn_data = load_energy_data('METADYN-1.ener')  # For METADYN (QM/MM) run

# Extract time (1st column) and potential energy (5th column) for each simulation
time_metadyn = metadyn_data[:, 0]
pot_energy_metadyn = metadyn_data[:, 4]


# Create 'Figures' directory if it doesn't exist
if not os.path.exists('Figures'):
    os.makedirs('Figures')

# Plot Potential Energy vs Time for METADYN (QM/MM), NVT, and NPT simulations
plt.figure(figsize=(10, 6))

plt.plot(time_metadyn, pot_energy_metadyn, label='METADYN (QM/MM)', color='green', linewidth=2)

# Add axis labels and title
plt.xlabel('Time (ps)', fontsize=12)
plt.ylabel('Potential Energy (kJ/mol)', fontsize=12)
plt.title('Potential Energy vs Time (QM/MM)', fontsize=14)

# Add legend
plt.legend(fontsize=12)

# Add grid for better readability
plt.grid(True)

# Save the plot
plt.savefig('Figures/potential_energy_vs_time_metadyn.png', dpi=300)

# Show the plot
plt.show()
```
