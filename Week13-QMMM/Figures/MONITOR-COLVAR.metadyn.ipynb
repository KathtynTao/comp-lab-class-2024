```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Function to load the reaction coordinate data
def load_reaction_coordinate(file_path):
    return np.loadtxt(file_path, comments='@')

# Load the reaction coordinate data from MONITOR-COLVAR.metadynLog
reaction_data = load_reaction_coordinate('MONITOR-COLVAR.metadynLog')

# Extract time (1st column) and reaction coordinate (2nd column)
time = reaction_data[:, 0]  # Time is typically in the first column
reaction_coord = reaction_data[:, 1]  # Reaction coordinate is typically in the second column

# Create 'Figures' directory if it doesn't exist
if not os.path.exists('Figures'):
    os.makedirs('Figures')

# Plot Reaction Coordinate vs Time
plt.figure(figsize=(10, 6))

plt.plot(time, reaction_coord, label='Reaction Coordinate', color='purple', linewidth=2)

# Add axis labels and title
plt.xlabel('Time (ps)', fontsize=12)
plt.ylabel('Reaction Coordinate', fontsize=12)
plt.title('Reaction Coordinate vs Time (QM/MM Simulation)', fontsize=14)

# Add grid for better readability
plt.grid(True)

# Save the plot
plt.savefig('Figures/reaction_coordinate_vs_time.png', dpi=300)

# Show the plot
plt.show()
```
