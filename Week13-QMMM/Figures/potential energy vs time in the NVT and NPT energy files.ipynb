```python
import numpy as np
import matplotlib.pyplot as plt
import os  # For checking and creating directories

# Load the potential energy data from the .ener or .xvg files
# Ensure the file paths are correct based on your system and file extensions

# For NVT simulation (adjust file name if needed)
nvt_data = np.loadtxt('NVT-1.ener', comments='@', skiprows=24)

# For NPT simulation (adjust file name if needed)
npt_data = np.loadtxt('NPT-1.ener', comments='@', skiprows=24)

# Extract time (1st column) and potential energy (5th column) for NVT and NPT
time_nvt = nvt_data[:, 0]  # Time in ps (1st column)
pot_energy_nvt = nvt_data[:, 4]  # Potential energy in kJ/mol (5th column)

time_npt = npt_data[:, 0]  # Time in ps (1st column)
pot_energy_npt = npt_data[:, 4]  # Potential energy in kJ/mol (5th column)

# Create the 'Figures' directory if it doesn't exist
if not os.path.exists('Figures'):
    os.makedirs('Figures')

# Create a plot
plt.figure(figsize=(10, 6))

# Plot NVT and NPT potential energy vs time
plt.plot(time_nvt, pot_energy_nvt, label='NVT', color='blue', linewidth=2)
plt.plot(time_npt, pot_energy_npt, label='NPT', color='red', linewidth=2)

# Add axis labels and title
plt.xlabel('Time (ps)', fontsize=12)
plt.ylabel('Potential Energy (kJ/mol)', fontsize=12)
plt.title('Potential Energy vs Time (NVT and NPT Simulations)', fontsize=14)

# Add a legend
plt.legend(fontsize=12)

# Add grid for better readability
plt.grid(True)

# Save the plot as a PNG file in the Figures directory
plt.savefig('Figures/potential_energy_vs_time.png', dpi=300)

# Show the plot
plt.show()
```
