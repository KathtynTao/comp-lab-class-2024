```python
file_path = 'Data/1hz3_T310.run.25000000.energy.xvg'

# Read the file into a pandas DataFrame
# The file format may include header information that we need to skip
# We'll assume that the file uses whitespace as delimiter and has a header that we want to skip
# You might need to adjust the header and delimiter according to your file specifics

# Function to read the data file
def read_data(file_path):
    # Read the file, assuming that the header starts after lines that begin with '@'
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    # Identify the line where data starts (typically after '@' lines)
    start_index = 0
    for i, line in enumerate(lines):
        if not line.startswith(('@', '#')):
            start_index = i
            break
    
    # Load the data into a DataFrame
    data = pd.read_csv(file_path, delim_whitespace=True, skiprows=start_index)
    
    return data

# Load the data
data = read_data(file_path)

# Display the first few rows to understand its structure
data.head()

# Define the time in nanoseconds (assuming the time column is the first column)
time_ns = data.iloc[:, 0] / 1000.0  # Assuming time is in picoseconds and we convert to nanoseconds

# Plot each quantity vs. time
quantities = data.columns[1:]  # Skip the time column

# Plot each quantity
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 10))
axes = axes.flatten()

for i, quantity in enumerate(quantities):
    axes[i].plot(time_ns, data[quantity])
    axes[i].set_xlabel('Time (ns)')
    axes[i].set_ylabel(quantity)
    axes[i].set_title(f'{quantity} vs Time')

plt.tight_layout()
plt.show()

# Plot kinetic energy and temperature on the same plot
# Assuming 'Kinetic' and 'Temperature' are the exact names of the columns
if 'Kinetic' in data.columns and 'Temperature' in data.columns:
    fig, ax1 = plt.subplots(figsize=(10, 6))
    
    color = 'tab:red'
    ax1.set_xlabel('Time (ns)')
    ax1.set_ylabel('Temperature', color=color)
    ax1.plot(time_ns, data['Temperature'], color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    
    ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis
    color = 'tab:blue'
    ax2.set_ylabel('Kinetic Energy', color=color)
    ax2.plot(time_ns, data['Kinetic'], color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    
    fig.tight_layout()  # To ensure the layout fits the figure area
    plt.title('Kinetic Energy and Temperature vs Time')
    plt.show()
else:
    print("Columns 'Kinetic' and/or 'Temperature' not found in the data.")
```

```python
# Ubiquitin Structure Analysis with MDTraj


# Import necessary libraries
import mdtraj as md

# Define the path to the ubiquitin structure file
# Make sure to replace 'path_to_ubiquitin_file' with the actual file path
file_path = '1UBQ_processed.pdb'

# Load the ubiquitin structure using MDTraj
ubiquitin = md.load(file_path)

# Print the total number of hydrogen bonds
# MDTraj uses a method to calculate hydrogen bonds
# We'll calculate hydrogen bonds with a default distance and angle threshold
h_bonds = md.wrapped_hbonds(ubiquitin, cutoff=0.35, angle=90.0)
num_h_bonds = len(h_bonds)
print(f"Total number of hydrogen bonds: {num_h_bonds}")

# Compute the number of helical amino acids
# We need to compute the secondary structure first
secondary_structure = md.compute_dssp(ubiquitin)
# Extract the secondary structure assignment
secondary_structure_str = secondary_structure[1]

# Count the number of helical residues
# 'H' represents an alpha helix in DSSP
num_helical_amino_acids = sum(residue == 'H' for residue in secondary_structure_str)
print(f"Number of helical amino acids: {num_helical_amino_acids}")

# Display the secondary structure assignments for verification
print("\nSecondary structure assignments:")
for i, sec_str in enumerate(secondary_structure_str):
    print(f"Residue {i+1}: {sec_str}")
```

```python
# Step 2: Load the MD Trajectory
# Replace '1hz3_T310.stepid25000000.every100ps.nowater.xtc' and '1hz3.pdb' with the correct filenames if different
trajectory_file = '1hz3_T310.stepid25000000.every100ps.nowater.xtc'
topology_file = '1hz3_T310.start.nowater.gro'  # Replace with the correct topology file

# Load the trajectory with the correct topology file
try:
    traj = md.load(trajectory_file, top=topology_file)
    print(f"Loaded trajectory with {traj.n_atoms} atoms and {traj.n_frames} frames.")
except Exception as e:
    print(f"An error occurred while loading the file: {e}")
```

```python
# Step 3: Compute the End-to-End Distance and Radius of Gyration at Each Time
# Select the atom indices of interest for end-to-end distance calculation
atom_indices = traj.top.select("name CA")  # Using all alpha carbons

# Calculate end-to-end distance (distance between first and last alpha carbons)
end_to_end_distances = md.compute_distances(traj, [[atom_indices[0], atom_indices[-1]]])[:, 0]

# Calculate radius of gyration
radii_of_gyration = md.compute_rg(traj)

# Check the results
print(f"Computed end-to-end distances for {len(end_to_end_distances)} frames.")
print(f"Computed radii of gyration for {len(radii_of_gyration)} frames.")
```

```python
# Step 4: Plot End-to-End Distance and Radius of Gyration vs Time
time = traj.time / 1000  # Convert time to nanoseconds if needed

plt.figure(figsize=(10, 6))
plt.plot(time, end_to_end_distances, label='End-to-End Distance (nm)', color='blue')
plt.plot(time, radii_of_gyration, label='Radius of Gyration (nm)', color='red')
plt.xlabel('Time (ns)')
plt.ylabel('Distance (nm)')
plt.title('End-to-End Distance and Radius of Gyration vs Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 5: Plot Normalized Histograms of End-to-End Distance and Radius of Gyration

# Plot normalized histogram of end-to-end distance
plt.figure(figsize=(10, 4))
plt.hist(end_to_end_distances, bins=30, density=True, alpha=0.7, color='blue')
plt.xlabel('End-to-End Distance (nm)')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of End-to-End Distance')
plt.grid(True)
plt.show()

# Plot normalized histogram of radius of gyration
plt.figure(figsize=(10, 4))
plt.hist(radii_of_gyration, bins=30, density=True, alpha=0.7, color='red')
plt.xlabel('Radius of Gyration (nm)')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of Radius of Gyration')
plt.grid(True)
plt.show()
```
