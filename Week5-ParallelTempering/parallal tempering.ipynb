```python
import numpy as np
import matplotlib.pyplot as plt

# Load energy data from the .xvg files
def load_energy_data(filename):
    data = np.loadtxt(filename, comments=['@', '#'])
    return data[:, 1]  # Assuming the potential energy is in the second column

# Load potential energy data
energy_T300 = load_energy_data('potential_energy_T300.xvg')
energy_T400 = load_energy_data('potential_energy_T400.xvg')
energy_T600 = load_energy_data('potential_energy_T600.xvg')

# Create histograms
plt.figure(figsize=(12, 6))
plt.hist(energy_T300, bins=30, alpha=0.5, label='T300', density=True)
plt.hist(energy_T400, bins=30, alpha=0.5, label='T400', density=True)
plt.hist(energy_T600, bins=30, alpha=0.5, label='T600', density=True)

# Plotting details
plt.title('Histogram of Potential Energy for Different Temperatures')
plt.xlabel('Potential Energy')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```
