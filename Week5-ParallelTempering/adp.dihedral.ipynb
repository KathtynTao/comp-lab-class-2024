```python
import mdtraj as md
import matplotlib.pyplot as plt

# Load the trajectory and topology file
trajectory = md.load('adp_exchange3temps.xtc', top='adp.gro')  # Adjust the filenames as necessary

# Compute phi and psi dihedral angles
phi_indices = md.compute_phi(trajectory)
psi_indices = md.compute_psi(trajectory)

# Extract the angles in radians
phi_angles = phi_indices[0] * (180 / np.pi)  # Convert to degrees
psi_angles = psi_indices[0] * (180 / np.pi)  # Convert to degrees

# Extract time information
time = trajectory.time  # Time in ps

# Plot phi vs time
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(time, phi_angles)
plt.title('Phi Dihedral Angles Over Time')
plt.xlabel('Time (ps)')
plt.ylabel('Phi Angle (degrees)')
plt.grid()

# Plot psi vs time
plt.subplot(2, 1, 2)
plt.plot(time, psi_angles)
plt.title('Psi Dihedral Angles Over Time')
plt.xlabel('Time (ps)')
plt.ylabel('Psi Angle (degrees)')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np

# Load the trajectory file (assumes GROMACS outputs .xtc or .trr file and corresponding topology file .gro)
trajectory = md.load('adp_exchange3temps.xtc', top='adp.gro')

# Compute the phi and psi dihedral angles for alanine dipeptide
phi_indices, phi_angles = md.compute_phi(trajectory)
psi_indices, psi_angles = md.compute_psi(trajectory)

# Convert time to nanoseconds (assuming 2 fs time step in the simulation)
time_ns = trajectory.time / 1000

# Plot phi vs time
plt.figure(figsize=(10, 6))
plt.plot(time_ns, np.rad2deg(phi_angles[:, 0]), label='Phi', color='blue')
plt.xlabel('Time (ns)')
plt.ylabel('Phi angle (degrees)')
plt.title('Phi Dihedral Angle vs Time')
plt.grid(True)
plt.legend()
plt.show()

# Plot psi vs time
plt.figure(figsize=(10, 6))
plt.plot(time_ns, np.rad2deg(psi_angles[:, 0]), label='Psi', color='green')
plt.xlabel('Time (ns)')
plt.ylabel('Psi angle (degrees)')
plt.title('Psi Dihedral Angle vs Time')
plt.grid(True)
plt.legend()
plt.show()
```
