```python
import matplotlib.pyplot as plt
import numpy as np

# Function to load the .xvg file and skip metadata lines starting with # or @
def load_xvg(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            if not line.startswith(('#', '@')):
                data.append([float(x) for x in line.split()])
    return np.array(data)

# Load the potential.xvg file
file_path = 'potential.xvg'
data = load_xvg(file_path)

# Assuming the first column is time and the second column is potential energy
time = data[:, 0]
potential_energy = data[:, 1]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(time, potential_energy, label='Potential Energy')

# Adding titles and labels
plt.title('Potential Energy vs Time')
plt.xlabel('Time (ps)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.legend()

# Show the plot
plt.show()
```
