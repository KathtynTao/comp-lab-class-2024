```python
import mdtraj as md

# Load the PDB file as a single frame
traj = md.load('6ZSL_ATP_RNA.pdb')

# Select phosphate atoms (P atoms in RNA residues)
rna_phosphate_atoms = traj.topology.select("element P and (resname A or resname U or resname G or resname C)")

# Select uracil oxygen atoms (common names for oxygens in uracil are O2 and O4)
uracil_oxygen_atoms = traj.topology.select("resname U and (name O2 or name O4)")

# Find amino acids within 0.5 nm of the RNA phosphate atoms
phosphate_neighbors = md.compute_neighbors(traj, 0.5, rna_phosphate_atoms, haystack_indices=None)

# Find amino acids within 0.5 nm of the uracil oxygen atoms
uracil_oxygen_neighbors = md.compute_neighbors(traj, 0.5, uracil_oxygen_atoms, haystack_indices=None)

# Get the unique residues for each set of neighbors
phosphate_residues = set([traj.topology.atom(i).residue for i in phosphate_neighbors[0]])
uracil_oxygen_residues = set([traj.topology.atom(i).residue for i in uracil_oxygen_neighbors[0]])

# Print out the residues near phosphate atoms
print("Amino acids within 0.5 nm of RNA phosphate atoms:")
for residue in phosphate_residues:
    print(residue)

# Print out the residues near uracil oxygen atoms
print("\nAmino acids within 0.5 nm of RNA uracil oxygen atoms:")
for residue in uracil_oxygen_residues:
    print(residue)
```
